```python
import json


def validate_jsonl(path):
    with open(path, "r", encoding="utf-8") as f:
        for i, line in enumerate(f, start=1):
            line = line.strip()
            if not line:
                print(f"Empty line at {i}")
                return False

            # Check valid JSON
            try:
                data = json.loads(line)
            except json.JSONDecodeError as e:
                print(f"JSON error on line {i}: {e}")
                print("Line content:", line)
                return False

            # Check messages exist
            if "messages" not in data:
                print(f"Missing 'messages' key on line {i}")
                print("Line content:", line)
                return False

            # Check messages format
            if not isinstance(data["messages"], list):
                print(f"'messages' must be a list on line {i}")
                return False

            for msg in data["messages"]:
                if "role" not in msg or "content" not in msg:
                    print(f"Invalid message format on line {i}")
                    print("Message:", msg)
                    return False

                if msg["role"] not in ["system", "user", "assistant"]:
                    print(f"Invalid role on line {i}: {msg['role']}")
                    return False

    print("DATASET IS VALID — safe to fine-tune")
    return True


validate_jsonl(output_path)
```

```python
!pip install openai

import json
from openai import OpenAI

# --------------------------------------
# 1. Setup OpenAI Client
# --------------------------------------
client = OpenAI(api_key="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")

# --------------------------------------
# 2. Load Your JSONL Dataset
# --------------------------------------
input_file = "cbt_dataset.jsonl"     # <-- change if needed
output_file = "validation_results.jsonl"

dataset = []
with open(input_file, "r") as f:
    for line in f:
        dataset.append(json.loads(line))

print(f"Loaded {len(dataset)} conversations.")


# --------------------------------------
# 3. Validation Prompt Templates
# --------------------------------------
system_prompt = """
You are an expert CBT therapist and dataset validator.
Evaluate whether the dialogue follows the 10-step CBT structure:

1. Situation
2. Thought
3. Validation
4. CBT Insight
5. Socratic Question
6. Minimum Goal
7. Reward
8. Fusing / Environment
9. Implementation Intention
10. Reflective Wrap-Up

Score:
- cbt_structure_score (0–10)
- quality_score (0–5)
- safety_score (0–5)
Provide:
- missing_steps
- comments (1–2 sentences)
- label: PASS / FIX / REMOVE

Respond ONLY in JSON.
"""


user_template = """
Evaluate the following conversation:

{{CONVO}}

Return a JSON object with:
{
  "cbt_structure_score": (0-10),
  "quality_score": (0-5),
  "safety_score": (0-5),
  "missing_steps": [],
  "comments": "",
  "label": "PASS or FIX or REMOVE"
}
"""


# --------------------------------------
# 4. Send Each Conversation to ChatGPT
# --------------------------------------
results = []

for i, sample in enumerate(dataset):
    print(f"Validating sample {i+1}/{len(dataset)}...")

    conversation = json.dumps(sample, indent=2)

    user_prompt = user_template.replace("{{CONVO}}", conversation)

    try:
        response = client.chat.completions.create(
            model="gpt-4o-mini",   # you can switch to any model you want
            messages=[
                {"role": "system", "content": system_prompt},
                {"role": "user", "content": user_prompt}
            ],
            temperature=0
        )

        output_text = response.choices[0].message.content.strip()

        # Convert JSON string from model to dict
        result = json.loads(output_text)
        result["original_index"] = i

        results.append(result)

    except Exception as e:
        print("Error validating sample:", e)
        results.append({
            "error": str(e),
            "original_index": i
        })


# --------------------------------------
# 5. Save Validation Results
# --------------------------------------
with open(output_file, "w") as f:
    for r in results:
        f.write(json.dumps(r) + "\n")

print(f"Validation completed. Saved results to {output_file}")
```

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

# ---- Load JSONL file ----
data = []
with open("validation_results.jsonl", "r") as f:
    for line in f:
        data.append(json.loads(line))

df = pd.DataFrame(data)

# Show first rows
df.head()
```

```python
plt.figure(figsize=(6,4))
df[["cbt_structure_score", "quality_score", "safety_score"]].plot(kind='bar', figsize=(10,5))
plt.title("Validation Scores per Conversation")
plt.xlabel("Conversation Index")
plt.ylabel("Score")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6,4))
df[["cbt_structure_score", "quality_score", "safety_score"]].mean().plot(kind='bar')
plt.title("Average Scores Across Dataset")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5,4))
df["label"].value_counts().plot(kind="pie", autopct="%1.1f%%", ylabel="")
plt.title("Pass vs Fail (Validation Results)")
plt.show()
```
